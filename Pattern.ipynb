```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
df1=pd.read_csv('adult.data',names=['Age','Workclass','FNLWGT','Education','Education-Num','Marital-Status','Occupation',
                               'Relationship','Race','Sex','Capital-Gain','Capital-Loss','Hours-per-Week','Native-Country',
                               'Income' ])
df1
```

```python
df2=pd.read_csv('adult.test',names=['Age','Workclass','FNLWGT','Education','Education-Num','Marital-Status','Occupation',
                               'Relationship','Race','Sex','Capital-Gain','Capital-Loss','Hours-per-Week','Native-Country',
                               'Income' ])
df2
```

```python
y_train= df1.iloc[:, -1]
print(y_train)

y_train = y_train.str.strip()
mask1 = y_train == '>50K'
y_train[mask1] = 'True'
y_train[~mask1] = 'False'
y_train=pd.DataFrame(y_train)
y_train
```

```python
df1_whithout_income=df1.iloc[:, :-1] 
df1_whithout_income
```

```python
y_test= df2.iloc[:, -1]
print(y_test)

y_test = y_test.str.strip()
mask2 = y_test == '>50K.'
y_test[mask2] = 'True'
y_test[~mask2] = 'False'
y_test=pd.DataFrame(y_test)
y_test
```

```python
df2_whithout_income=df2.iloc[:, :-1] 
df2_whithout_income
```

```python
concatenated_df = pd.concat([df1_whithout_income, df2_whithout_income],axis=0,ignore_index=True)
concatenated_df 
```

```python
categorical_data=concatenated_df.select_dtypes(exclude=[np.number]).columns
numerical_data=concatenated_df.select_dtypes(include=[np.number]).columns
encoder = OneHotEncoder(drop='first', sparse=False)
encoded_data = encoder.fit_transform(concatenated_df[categorical_data])
encoded_data_df = pd.DataFrame(encoded_data)
data_encoded = pd.concat([concatenated_df[numerical_data], encoded_data_df],axis=1)
data_encoded
```

```python
x_train = data_encoded.iloc[:32561]
x_train.columns = x_train.columns.astype(str)
x_train
```

```python
x_test = data_encoded.iloc[32561:]
x_test.columns = x_test.columns.astype(str)
x_test
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_test
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
model.predict_proba(x_test)
```

```python
TN, FP,FN,TP= confusion_matrix(y_test,y_predict).ravel()
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
accuracy = (TP + TN) / (TP + TN + FP + FN)

print("Sensitivity :", sensitivity)
print("Specificity :", specificity)
print("accuracy :", accuracy)
```
